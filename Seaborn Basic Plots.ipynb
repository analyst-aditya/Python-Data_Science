1. Scatter Plot

👉 Do numerical variables ke beech relationship dekhne ke liye.

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")

sns.scatterplot(x="total_bill", y="tip", data=tips)
plt.show()
```

2. Line Plot

👉 Time series ya trend dikhane ke liye.

```python
sns.lineplot(x="size", y="total_bill", data=tips)
plt.show()
```

3. Bar Plot

👉 Categories ka average value dikhata hai.

```python
sns.barplot(x="day", y="total_bill", data=tips)
plt.show()
```

4. Count Plot

👉 Sirf frequency (kitne values hain) dikhata hai.

```python
sns.countplot(x="day", data=tips)
plt.show()
```

5. Histogram / Distribution Plot

👉 Numerical column ki distribution dekhne ke liye.

```python
sns.histplot(tips["total_bill"], bins=20, kde=True)
plt.show()
```

```python
# ⚡ Summary of Step 3:

# scatterplot() → Relationship between 2 numbers

# lineplot() → Trend/time series

# barplot() → Average comparison

# countplot() → Frequency counts

# histplot() → Distribution
```
