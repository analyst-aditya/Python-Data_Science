```python
import plotly.express as px
import seaborn as sns
```

```python
df = sns.load_dataset('car_crashes')    
df.head()
```

```python
fig= px.bar(df,x='abbrev',y='total',
            title='total car Accidents by state',
            labels={'abbrev':'state','total':'total accidents'},
)
fig.show()
```

```python
# importing plotly.io to save the figure as an image where io stands for input/output
import plotly.io as pio 
pio.write_image(fig, 'total_car_accidents_by_state.png') # saving the figure as
```

```python
df = sns.load_dataset('car_crashes')
df.head()
```

```python
top10 = df.sort_values('ins_premium', ascending=False).head(10)
fig= px.bar(top10,x='abbrev',y='ins_premium',
            title='top 10 states with highest insurance premium',
            labels={'abbrev':'state','ins_premium':'insurance premium'},
            color = 'ins_premium',
            color_continuous_scale=px.colors.sequential.Agsunset,
            template='plotly_dark'
)
fig.show()
```

```python
fig =px.pie(df.head(10),values = 'speeding', names ='abbrev',
         title='car accidents due to speeding accidents by statee',
         color_discrete_sequence=px.colors.sequential.Darkmint_r,
         template='plotly_dark'
)
fig.update_traces(textposition='inside')
fig.show()
```

```python
ig = px.scatter(df,x = 'alcohol', y = 'speeding',
                title = 'alcohol vs speeding',
                labels = {'alcohol':'alcohol related accidents',
                          'speeding':'speeding related accidents'}, 
                          color='abbrev',
                          color_discrete_sequence=px.colors.sequential.Agsunset, 
                          template='plotly_dark', 
                          size='total'
)
ig.update_layout(title ={'x':0.5})
ig.show()
```

```python
import plotly.express as px
import seaborn as sns

# Load taxis dataset
data = sns.load_dataset("taxis")

# Scatter plot
fig = px.scatter(
    data,
    x="distance",
    y="fare",
    color="passengers",   # color points by number of passengers
    template='plotly_dark',     # add dark theme
    size="tip",           # bubble size = tip amount
    title="Taxi Fare vs Distance (Bubble size = Tip, Color = Passengers)",
    labels={"fare": "Fare ($)", "distance": "Distance (miles)"}
)

fig.show()
```

```python
fig = px.histogram(
    data,
    x="fare",
    nbins=50,
    title="Distribution of Taxi Fares",
    template='plotly_dark', 
    labels={"fare": "Fare ($)"},
    color_discrete_sequence=px.colors.sequential.Agsunset
    
)
fig.show()
```

```python
fig = px.box(
    data,
    x="pickup_zone",
    y="tip",
    title="Tip Distribution by Pickup Zone",
    template='plotly_dark', 
    labels={"pickup_zone": "Pickup Zone", "tip": "Tip ($)"}
    
)
fig.show()
```

```python
import plotly.express as px
import seaborn as sns

# Load taxis dataset
data = sns.load_dataset("taxis")

# Pie chart of payment types
fig = px.pie(
    data,
    names="payment",   # column for categories
    template='plotly_dark', 
    title="Distribution of Payment Methods in Taxi Rides"
)

fig.show()
```

```python
fig = px.pie(
    data,
    names="passengers",  # group by number of passengers
    template='plotly_dark', 
    title="Share of Taxi Rides by Number of Passengers"
    
)
fig.show()
```

```python
# 3d scatter plot
fig = px.scatter_3d(
    data,
    x="distance",
    y="fare",
    z="tip",
    color="passengers",   # color points by number of passengers
    template='plotly_dark',     # add dark theme
    size="tip",           # bubble size = tip amount
    title="3D Scatter of Fare vs Distance vs Tip",
    labels={"fare": "Fare ($)", "distance": "Distance (miles)", "tip": "Tip ($)"}
)   
fig.show()    
```
