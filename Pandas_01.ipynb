## Grouping and aggeresion in pandas 

#### __groupby() basic__ --> Kisi column ke basis par data ko group karna 

```python
# groupby() basic 
import pandas as pd

data = {
    "name": ['a','b','c','d'],
    "salary": [20000,21000,19000,18000]
}
fd = pd.DataFrame(data)
print(fd.groupby('name')['salary'].mean())
```

#### __Aggregation (mean, sum, count...etc)__ --> Group ke uper summary function lagana

```python
fd.groupby('name')['salary'].mean().max()
```

```python
fd.groupby('name')['salary'].mean().count()
```

```python
fd.groupby('name')['salary'].mean().sum()
```

#### __agg()__ for multiple function apply karne ke liye
- ek hi column par multiple function apply karna

```python
fd.groupby('name')['salary'].agg(['mean','sum','max'])
```

#### Multi-level grouping
- ek se zyada columns ke basis par grouping

```python
import pandas as pd

data = {
    "name": ['a','b','c','d'],
    'gender': ['M','F','M','M'],
    "salary": [20000,21000,19000,18000]
}
cd = pd.DataFrame(data)
print(cd.groupby(['name','gender'])['salary'].mean())
```

#### Pivot tables & crosstab
- Excel jaise sumary table banane ke liye 

__'_Pivot table_'__

```python
# Pivot table
print(cd.pivot_table(values='salary', index='name', columns='gender',aggfunc="mean"))
```

__'_Crosstab ( frequency count )_'__

```python
print(pd.crosstab(cd['name'],cd['gender']))
```

## Merging & Joining in Pandas

#### Concaenation --> pd.concat()
- DataFrame ko row wise ya colunm wise jodna

```python
import pandas as pd

df1 = pd.DataFrame({"ID": [1, 2], "Name": ["A", "B"]})
df2 = pd.DataFrame({"ID": [3, 4], "Name": ["C", "D"]})

# Row-wise concat
print(pd.concat([df1, df2]))

# Column-wise concat
print(pd.concat([df1, df2], axis=1))
```

#### Merge --> pd.merge()
- SQL JOIN ki tarah kamm karta hai (common column ke basis par)

```python
employees = pd.DataFrame({
    "ID": [1, 2, 3],
    "Name": ["A", "B", "C"]
})

salary = pd.DataFrame({
    "ID": [1, 2, 4],
    "Salary": [50000, 60000, 70000]
})

# Default: INNER JOIN
print(pd.merge(employees, salary, on="ID"))

# LEFT JOIN
print(pd.merge(employees, salary, on="ID", how="left"))

# RIGHT JOIN
print(pd.merge(employees, salary, on="ID", how="right"))

# OUTER JOIN
print(pd.merge(employees, salary, on="ID", how="outer"))
```

➡ Types of Joins:

INNER JOIN → Common values only

LEFT JOIN → Left DataFrame ke sab + matching right

RIGHT JOIN → Right DataFrame ke sab + matching left

OUTER JOIN → Dono ke sab values (missing = NaN)

#### ✅ (3) Join → df.join()

- 👉 Mostly index ke basis par join karta hai.

```python

df3 = pd.DataFrame({"Name": ["A", "B", "C"]}, index=[1,2,3])
df4 = pd.DataFrame({"Salary": [50000, 60000, 70000]}, index=[1,2,3])

print(df3.join(df4))
```

#### ✅ (4) One-to-One, One-to-Many, Many-to-Many

- One-to-One → Ek row match karta hai dusre ek row se.

- One-to-Many → Ek row match hota hai multiple rows se.

- Many-to-Many → Dono taraf multiple matches ho sakte hain.



#### ⚡Short Summary:

- concat() → DataFrames ko row/column wise jodna

- merge() → SQL join jaisa (on = column)

- join() → Mostly index ke basis par join

- Supports INNER, LEFT, RIGHT, OUTER joins
