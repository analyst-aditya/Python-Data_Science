```python
# if-else statement

a = 21
if a >= 18 :
    print("You are elegible for vote")
else : 
    print("You are not eligible for vote")
```

```python
# if-elif-else statement

b = 0

if b > 0:
    print("The value is positive")
elif b == 0 :
    print("The value is zero")
else:
    print("The value is negative")
```

```python
# nested if-else statement

num = 56

if num < 0:
    print("num is negative")
elif num > 0:
    if num <= 10:
        print("num is between 1-10")
    elif num > 10 and num < 20:
        print("num is between 11-20")
    else:
        print("num is greater than 20")
else:
    print("num is zero")
```

```python
# nested-if example

username = input("Enter your username: ")
email = input("Enter your email: ")
password = input("Enter your password: ")
cpassword = input("Enter your confirm password: ")
if len(username) > 5 and len(username)<=15:
    if "@" in email and len(email) > 10:
        if password != username and len(password) >= 8:
            if password == cpassword:
                print("Registration successful")
            else:
                print("Password mismatch")
        else:
            print("invalid password")
    else:
        print("Invalid email format")    
else:
    print("invalid username")
```
