# Grouping and aggeresion in pandas 

#### __groupby() basic__ --> Kisi column ke basis par data ko group karna 

```python
# groupby() basic 
import pandas as pd

data = {
    "name": ['a','b','c','d'],
    "salary": [20000,21000,19000,18000]
}
fd = pd.DataFrame(data)     # it convert the dictionary into dataframe
print(fd.groupby('name')['salary'].mean())     

# fd.groupby('name'): it group the data on the basis of name column, means if same name use many time so that makes group
# ['salary'].mean(): for grouping takes only salary column, then find average salary of each group

```

```python
val = { "name": ['a','b','a','b'],
        "salary": [2000,3000,4000,5000]}
kd = pd.DataFrame(val)     
print(kd.groupby('name')['salary'].mean())     # yaha par name ke according group kiya gya hi same namehone par unki salary ka average nikala gya hai 
ck = kd.groupby('name')['salary'].mean().max()
print(ck)
```

#### __Aggregation (mean, sum, count...etc)__ --> Group ke uper summary function lagana

```python
ck = fd.groupby('name')['salary'].mean().max()          # ye line same name par max value return karti hai
print(ck)
```

```python
cnt = fd.groupby('name')['salary'].mean().count()   # isme count ka use karke hum group ke andar kitne unique name hai wo count kar sakte hai
print(cnt)
```

```python
sm = fd.groupby('name')['salary'].mean().sum()
print(sm)
```

#### __agg()__ for multiple function apply karne ke liye
- ek hi column par multiple function apply karna

```python
agg = fd.groupby('name')['salary'].agg(['mean','sum','max'])
print(agg)
```

#### Multi-level grouping
- ek se zyada columns ke basis par grouping

```python
import pandas as pd

data = {
    "name": ['a','b','c','d'],
    'gender': ['M','F','M','M'],
    "salary": [20000,21000,19000,18000]
}

cd = pd.DataFrame(data)
print(cd.groupby(['name','gender'])['salary'].mean())
```

#### Pivot tables & crosstab
- Excel jaise sumary table banane ke liye 

__'_Pivot table_'__

```python
# Pivot table
print(cd.pivot_table(values='salary', index='name', columns='gender',aggfunc="mean"))
```

__'_Crosstab ( frequency count )_'__

```python
print(pd.crosstab(cd['name'],cd['gender'])) # crosstab ka use ya jyada variables ke beech relationship ya frequency distibution ko table form me analyze karne ke liye kita jata h.
```
