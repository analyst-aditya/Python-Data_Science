```python
import plotly.express as px
import seaborn as sns
```

```python
df = sns.load_dataset('car_crashes')
df.head()
```

```python
import plotly.express as px

fig = px.bar(df, x='abbrev', y='total',
             title='Bar plot of total by abbrev',
             labels={'abbrev': 'State', 'total': 'Total accidents'})
fig.show()

```
