# 🎯 Assignment 2 Summary

Tumne seekha:

✔ Missing values detect
✔ Text column clean
✔ Numeric column clean
✔ Duplicate remove

## City Column Clean Karna

🔎 Explanation

👉 mode() → most common city

👉 [0] → first value

👉 fillna() → missing value replace karta hai

👉 inplace=True → original dataframe update karega

````python
df["City"].fillna(df["City"].mode()[0], inplace=True)       # mode() → most common city, [0] → first value, fillna() → missing value replace karta hai, inplace=True → original dataframe update karega
print(df.head(5))
````

## Profit Column Clean Karna

Profit numeric column hai.

Best practice:
👉 Mean ya Median se fill karte hain.

🔎 Explanation

👉 mean() → average profit

👉 Missing profit values ko average se replace kar diya

````python
df["Profit"].fillna(df["Profit"].mean(), inplace=True)
print(df.head(5))
````

````python
df.duplicated().sum()       # np.int64(0) iska matlab hai ki koi duplicate rows nahi hai aur agar hoti to np.int64(1) ya usse zyada hoti to iska matlab hota ki duplicate rows hai.
````

````python
df.drop_duplicates(inplace=True)    # Duplicate Remove (Agar Mile)
````

# 🚀 Assgnment 5 : Pivote Tables

### Task 1: City VS Category Sales Report

````python
pivot_month_product = pd.pivot_table(
    df,
    values="Sales",
    index="Month",
    columns="Product",
    aggfunc="sum"
)
print(pivot_month_product)
````

### TASK 3 — Multiple Values Pivot

````python
multi_pivot = pd.pivot_table(
    df,
    values=["Sales", "Profit"],
    index="Category",
    aggfunc=["sum", "mean"]
)

print(multi_pivot)
````

# 🚀 Assignment 6 — Feature Engineering

## 🧠 Feature Engineering Kya Hota Hai?

```
Existing data se new useful columns banana
```

Example:

* Profit Margin
* Order Size Category
* Customer Type
* Sales per Unit



### TASK 1 — Profit Margin Column Banana 
Formula:

```
Profit Margin = Profit / Sales
```

````python
df["Profit_Margin"] = df["Profit"] / df["Sales"]

print(df[["Sales", "Profit", "Profit_Margin"]].head())
````

````python
df.columns
````

### TASK 2 — Sales per Unit
Formula:

```
Sales per Unit = Sales / Quantity
```

````python
df["Sales_per_Unit"] = df["Sales"] / df["Quantity"]

print(df[["Sales", "Quantity", "Sales_per_Unit"]].head())
````

### TASK 3 — Order Value Category 

Hum order ko classify karenge:

| Sales     | Category |
| --------- | -------- |
| <1500     | Low      |
| 1500–3500 | Medium   |
| >3500     | High     |


````python
def order_category(sales):
    if sales < 1500:
        return "Low"
    elif sales <= 3500:
        return "Medium"
    else:
        return "High"

df["Order_Category"] = df["Sales"].apply(order_category)        # apply() use karke har row par logic apply kiya

print(df[["Sales", "Order_Category"]].head())
````

### TASK 4 — Customer Total Spending

````python
customer_total = df.groupby("Customer_Name")["Sales"].transform("sum")          # transform("sum"): Har customer ka total spending har row me add ho gaya

df["Customer_Total_Spending"] = customer_total

print(df[["Customer_Name", "Customer_Total_Spending"]].head())
````

### TASK 5 — Profit Type (Loss / Gain)


````python
df["Profit_Type"] = df["Profit"].apply(lambda x: "Loss" if x < 0 else "Gain")

print(df[["Profit", "Profit_Type"]].head())
````

# 🚀 Assignment 7 — Time Series Analytics

### STEP 1 — Date Column Ko Proper Datetime Banana


````python
df["Order_Date"] = pd.to_datetime(df["Order_Date"])     # Date ko proper datetime format me convert kiya

print(df["Order_Date"].dtype)
````

### STEP 2 — Year, Month, Day Extract Karna

````python
df["Year"] = df["Order_Date"].dt.year
df["Month"] = df["Order_Date"].dt.month
df["Day"] = df["Order_Date"].dt.day
df["Weekday"] = df["Order_Date"].dt.day_name()

print(df[["Order_Date","Year","Month","Weekday"]].head())
````

### STEP 3 — Monthly Sales Trend 

````python
monthly_sales = df.groupby("Month")["Sales"].sum().reset_index()
print(monthly_sales)
````

### STEP 4 — Year Wise Profit Growth

````python
yearly_profit = df.groupby("Year")["Profit"].sum().reset_index()

print(yearly_profit)
````

### STEP 5 — Weekday vs Weekend Analysis

````python
df["Is_Weekend"] = df["Weekday"].isin(["Saturday","Sunday"])

weekend_analysis = df.groupby("Is_Weekend")["Sales"].mean().reset_index()

print(weekend_analysis)
````

### STEP 6 — Resampling (Advanced Time Series)

````python
df.set_index("Order_Date", inplace=True)

monthly_resample = df["Sales"].resample("M").sum()      # `resample("M")` ➡ Monthly grouping karta hai


print(monthly_resample.head())
````

| Code | Meaning |
| ---- | ------- |
| "D"  | Daily   |
| "W"  | Weekly  |
| "M"  | Monthly |
| "Y"  | Yearly  |

# 🚀 Assignment 8 — Window Functions

## 🧠 Window Function Kya Hoti Hai?

Simple language:

```
Previous rows ko use karke calculation karna
```

Example:

👉 Running total
👉 Moving average
👉 Cumulative profit
👉 Ranking

---


### TASK 1 — Running Total Sales

````python
df["Running_Sales"] = df["Sales"].cumsum()      

print(df[["Sales", "Running_Sales"]].head())
````

👉 `cumsum()`
➡ Cumulative sum banata hai

Example:

```
Sales → 100
Sales → 200
Sales → 300
```

Running total:

```
100
300
600
```

---

### TASK 2 — Cumulative Profit

````python
df["Running_Profit"] = df["Profit"].cumsum()

print(df[["Profit", "Running_Profit"]].head())
````

### TASK 3 — Rolling Moving Average

````python
df["Rolling_Sales_3"] = df["Sales"].rolling(3).mean()       # Last 3 rows ka average Trend analysis me use hota hai


print(df[["Sales", "Rolling_Sales_3"]].head())
````

### TASK 4 — Expanding Window

````python
df["Expanding_Sales"] = df["Sales"].expanding().mean()

print(df[["Sales", "Expanding_Sales"]].head())
````

### TASK 5 — Ranking

````python
df["Sales_Rank"] = df["Sales"].rank(ascending=False)

print(df[["Sales", "Sales_Rank"]].head())
````

### TASK 6 — Group Wise Running Total

````python
df["Category_Running_Sales"] = df.groupby("Category")["Sales"].cumsum()

print(df[["Category", "Sales", "Category_Running_Sales"]].head())
````
