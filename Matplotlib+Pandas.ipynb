##  Visualization with Pandas

- Pandas me built-in plotting functions hote hain jo Matplotlib ke upar kaam karte hain. Ye quick data visualization ke liye best hote hain.

__⚡Short Summary:__

- .plot(kind="...") se line, bar, hist, scatter, pie, box ban sakte hain

- Multiple columns ek sath plot kar sakte ho

- Title, labels, colors, markers customize kar sakte ho

- Advanced graphs ke liye Pandas → Seaborn/Matplotlib


#### ✅ (1) Basic Plot

- 👉 DataFrame ya Series par .plot() lagakar simple graph bana sakte ho.


```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.DataFrame({
    "Month": ["Jan", "Feb", "Mar", "Apr"],
    "Sales": [200, 250, 300, 290]
})
print(df)

df.plot(x="Month", y="Sales", kind="line")
plt.show()

# ➡ Line chart (default) ban jayega.
```


#### ✅ (2) Common Plot Types

- Line plot → kind="line"

- Bar plot → kind="bar" / kind="barh"

- Histogram → kind="hist"

- Box plot → kind="box"

- Area plot → kind="area"

- Scatter plot → kind="scatter", x="col1", y="col2"

- Pie chart → kind="pie"


```python
# Example:

# Bar plot
df.plot(x="Month", y="Sales", kind="bar")

# Histogram
df["Sales"].plot(kind="hist")

# Scatter plot
df.plot(kind="scatter", x="Month", y="Sales")  # if numeric

```

#### ✅ (3) Multiple Columns Plot

- 👉 Ek sath multiple columns plot karna.

```python
df = pd.DataFrame({
    "Month": ["Jan", "Feb", "Mar", "Apr"],
    "Product_A": [200, 220, 250, 270],
    "Product_B": [150, 180, 210, 260]
})

df.plot(x="Month", y=["Product_A", "Product_B"], kind="line")
plt.show()

# ➡ Dono products ke trends ek hi graph me.
```


✅ (4) Customization

👉 Title, labels, grid, legend add kar sakte ho.

ax = df.plot(x="Month", y="Sales", kind="line", color="green", marker="o")
ax.set_title("Monthly Sales")
ax.set_xlabel("Month")
ax.set_ylabel("Sales")
plt.grid(True)
plt.show()


```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.DataFrame({
    "Month": ["Jan", "Feb", "Mar", "Apr"],
    "Product_A": [200, 220, 250, 270],
    "Product_B": [150, 180, 210, 260]
})

ax = df.plot(x="Month", y=["Product_A", "Product_B"], kind="line", marker="o")
ax.set_title("Monthly Sales")
ax.set_xlabel("Month")
ax.set_ylabel("Sales")
plt.grid(True)
plt.show()
```

#### ✅ (5) Pandas + Seaborn/Matplotlib

- 👉 Pandas ki plotting quick hoti hai, lekin advanced visualization ke liye Seaborn ya Matplotlib use karte hain. Pandas ke DataFrame ko directly in libraries me pass kar sakte ho.


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.DataFrame({
    "Month": ["Jan", "Feb", "Mar", "Apr"],
    "Sales": [200, 250, 300, 290]
})

sns.barplot(x="Month", y="Sales", data=df)
plt.title("Monthly Sales")         # Optional: Add title
plt.xlabel("Month")                # Optional: Add x label
plt.ylabel("Sales")                # Optional: Add y label
plt.show()
```
