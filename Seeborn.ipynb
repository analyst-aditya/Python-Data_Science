# Seaborn

- Integration with Pandas → directly pandas DataFrame ke sath kaam karta hai.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Built-in dataset load karna
tips = sns.load_dataset("tips")

# Simple scatterplot
sns.scatterplot(x="total_bill", y="tip", data=tips)

plt.show()
```

## 🎨 Seaborn Styling & Themes

### 1. set_style()

- Ye function plot ki background aur grid style change karta hai.

 __Available styles:__
- "white"

- "dark"

- "whitegrid" (default for many plots)

- "darkgrid"

- "ticks"

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_style("whitegrid")   # Background with white + grid

tips = sns.load_dataset("tips")
sns.scatterplot(x="total_bill", y="tip", data=tips)

plt.show()
```

### 2. set_palette()

- Ye function colorscheme set karta hai.
- Available palettes: "deep", "muted", "pastel", "bright", "dark", "colorblind"

```python
sns.set_palette("dark")        # set_palette() change the colore of plot
sns.countplot(x="day", data=tips)
plt.show()
```

### 3. set_context()

- Isse plot ka scale (size, font, line thickness) change hota hai.

__Contexts:__

- "paper" → small (research paper ke liye)

- "notebook" → medium (default)

- "talk" → large (presentation ke liye)

- "poster" → extra-large (big screen ke liye)

```python
sns.set_context("talk")
sns.barplot(x="day", y="total_bill", data=tips)
plt.show()
```

### 4. Combine Theme + Palette + Context

- Aap in sabko ek sath use kar sakte ho:

```python
sns.set_style("darkgrid")   # add background theme
sns.set_palette("bright")   # set color palette
sns.set_context("poster")   # size & fonts

sns.boxplot(x="day", y="tip", data=tips)
plt.show()
```
