```python
# Integer (integer is used to store whole numbers)
# It is used to store whole numbers, both positive and negative.
# It does not allow decimal points.
# It is defined by the absence of a decimal point.
age = 12
print("Age is : ", age)
print(type(age))
```

```python
# float(decimal value store krne k liye)
# It is used to store decimal numbers.
# It can represent both whole numbers and fractions.
# It is defined by a decimal point.

temperature = 23.34
print("Temperature: ", temperature)
print(type(temperature))
```

```python
# String (string is used to store text)
# It is a sequence of characters enclosed in single or double quotes.
# It can contain letters, numbers, and special characters.

name = "anuj kumar"
print(name)
print(type(name))
```

```python
# boolean (true or false)
# It is used to store binary values, either True or False.
# It is often used in conditional statements to control the flow of the program.
# In Python, boolean values are case-sensitive and must be capitalized as True or False.
is_student = True
print("Is Student: ", is_student)
print(type(is_student))
```

```python
# None ( jab kisi variable ki value abhi decide nhi ki h)
# None is a special constant in Python that represents the absence of a value or a null value.
# It is often used to signify that a variable has no value assigned to it.   

Homework_done = None
print("Homework_done: ", Homework_done)
print(type(Homework_done))
```

```python
# list(list ek multiple value store krti h)
# It is ordered, changeable, and allows duplicate values.
# It is defined by square brackets [].
# It can contain different data types, including strings, integers, and even other lists.
list = ["mango", "banana", "papaya", 5664]
print(list)
print(type(list))
```

```python
# tuple(ek baar value assign krne k baad change nhi hoti)
# tuple is a collection which is ordered and unchangeable. In Python, tuples are written with round brackets.
# It is used to store multiple items in a single variable.
# Tuples are similar to lists, but unlike lists, tuples cannot be changed or modified after creation.
# They are immutable.

tuple = ("mango", "banana", "papaya", 63)
print(tuple)
print(type(tuple))

colour = ("red", "blue", "white")
print(colour)
print(type(colour))
```

```python

# Dictionary-----> Dictionaries are used to store data values in key:value pairs.
# A dictionary is a collection which is ordered*, changeable and do not allow duplicates.

thisdict = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}
print("The dictionary is:", thisdict)
print(type(thisdict))
```

```python
# Set ---> A set is a collection which is unordered, unchangeable*, and unindexed. In Python, sets are written with curly brackets.
# Set is used to store unique values, and it does not allow duplicates.
# It is unordered, meaning the items do not have a defined order, and you cannot access items by index.
# Sets are mutable, meaning you can add or remove items after the set has been created.
Unique_numbers = {1, 0, 8, 9, 7, 3, 4, 3, 5 , 'fmv','fmv'}
print("Unique numbers are:", Unique_numbers)
print(type(Unique_numbers))
```
