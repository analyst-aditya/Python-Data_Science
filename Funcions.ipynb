#### Function Parameters

1. Positional Arguments ----> Jo order me diye jaate hain, function usi order me accept karta hai.

2. Keyword Arguments ----> Aap argument ka naam likh ke value de sakte ho, order matter nahi karta.

3. Default Arguments ----> Agar user koi value nahi deta to default use ho jaata hai.

4. Variable-Length Arguments ---->  *args → Multiple positional arguments

```python
# Positional Arguments
def greet(name, age):
    print("Name:", name)
    print("Age:", age)

greet("Aditya", 20)
print("-----------------------------")

# default argument
def greet(name="Guest"):
    print("Hello,", name)
greet()           # Default value lagegi
greet("Aditya")   # Argument diya to wahi use hoga

print("-----------------------------")

# Keyword Arguments ---> greet(age=20, name="Aditya")

# Variable Length Arguments
def total_marks(*marks):
    print("Marks:", marks)
    print("Total:", sum(marks))

total_marks(85, 90, 78)


```

```python
# Recursion function ---> function call itself
def factorial(n):
    if(n==1 or n==0):
        return 1
    return n * factorial(n-1)
n = int(input("Enter a number: "))
print(f"The factorial of this number is : {factorial(n)}")
```

 #### Scope and Lifetime of Variables in Python Functions

🔹 Scope = Variable kaha se access ho sakta hai </br>
🔹 Lifetime = Variable kab tak memory me rehta hai

#### Types of Scope in Python:
1. Local Scope --->
Jo variable function ke andar banaya gaya ho, wo sirf usi function ke andar accessible hota hai.

2. Global Scope --->
Jo variable function ke bahar banaya gaya ho, use pure program me kahin bhi access kiya ja sakta hai

3. global Keyword --->
Agar aap function ke andar global variable ko modify karna chahte ho, to global keyword ka use karna padega.

4. nonlocal Keyword --->
Function ke andar ek aur nested function ho, to inner function outer function ke variable ko nonlocal se modify kar sakta hai.

```python
# Local scope
def my_func():
    x = 10  # local variable
    print("Inside function:", x)
my_func()
print(x)  # ❌ Error: x is not defined outside
print("-----------------------------")

# global Keyword
x = 10      # global variable
def change():
    global x    # declare x as global isse x ko change kar sakte hain
    x = 20
change()
print(x)  # Output: 20
print("-----------------------------")


#  nonlocal Keyword
def outer():
    x = "hello"
    def inner():
        nonlocal x
        x = "world"
    inner()
    print(x)
outer()  # Output: world
print("-----------------------------")



```

#### Lambda Functions in Python (Anonymous Functions)
🔹 Lambda function ek single-line anonymous function hota hai — iska naam nahi hota</br>
🔹 Simple, short, one-time-use function ke liye use kiya jaata hai

```python
# lambda arguments: expression

# def square(x):
#     return x * x
# square(5)

# OR

square_lambda = lambda a: a * a  # Lambda function for square, isme lambda x: x * x likha hai jiska matlab hai ki x ko square karna hai
print("Square of 5 using lambda:", square_lambda(4))  # Output: 25
print("-----------------------------")

add = lambda x, y: x + y
print(add(5, 3))  # Output: 8
print("-----------------------------")

# Lambda inside map()
nums = [1, 2, 3, 4]
squares = list(map(lambda x: x ** 2, nums))
print(squares)  # Output: [1, 4, 9, 16]
print("-----------------------------")

# Lambda inside filter()
nums = [1, 2, 3, 4, 5]
even = list(filter(lambda x: x % 2 == 0, nums))
print(even)  # Output: [2, 4]

print("-----------------------------")

# Lambda inside sorted() with custom key
students = [("Aditya", 90), ("Raj", 85), ("Neha", 95)]
sorted_students = sorted(students, key=lambda x: x[1])
print(sorted_students)
# Output: [('Raj', 85), ('Aditya', 90), ('Neha', 95)]

```

```python
# find odd or even
odd_even = lambda n : "Even" if n%2 == 0 else "Odd"
print(odd_even(7))
```

```python
# WAP to calulate the area of a circle using Lmabda experession
area_circle = lambda r : 3.14 * (r**2)
print(area_circle(5))
```

## Generator
- Generator special type ka iterator hota hai jo automatically __ iter_() aur __ next_() implement karta hai
- generator function banane ke liye ham yield keyword ka use karte hai
- function ke yield pe pahuchne se function pause ho jata hai aur value return karta hai

```python
def my_generator():
    yield 1
    yield 2
    yield 3
gen = my_generator()
print(next(gen))
print(next(gen))
print(next(gen))
# print(next(gen))  # Error: StopIteration
```

```python
def even_nums():
    n = 0
    while True:
        yield n 
        n += 2
even = even_nums()
print(next(even))
print(next(even))
print(next(even))
print(next(even))
print(next(even))
```
