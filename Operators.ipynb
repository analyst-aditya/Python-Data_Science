# Operators

- Arithmatic Operator 
    - addition (+)
    - substraction(-)
    - multiplication(*)
    - division(/)
    - modulus(%)
    - Floor Division(//)
    - Exponentiation(**)


```python
a = 20
b = 5
print(a+b)  # Addition
print(a-b)  # Substraction
print(a*b)  # Multiplication
print(a/b)  # Division
print(a%b)  # Modulous
print(a**b) # Exponential
```

- Assignment Operators
    = ( x = 5 )    
    +=  (x += 5)    
    -= ( x -= 5)  
    *= (x *= 5)
    , /= (x /= 5)  
    //= (x //= 5)    
    **= ( x **= 5 )


```python
# Assignment Operators Examples

# 1. =
a = 10
print("a =", a)  # Output: 10

# 2. +=
a += 5   # a = a + 5
print("a += 5 →", a)  # Output: 15

# 3. -=
a -= 3   # a = a - 3
print("a -= 3 →", a)  # Output: 12

# 4. *=
a *= 2   # a = a * 2
print("a *= 2 →", a)  # Output: 24

# 5. /=
a /= 4   # a = a / 4
print("a /= 4 →", a)  # Output: 6.0 (Note: becomes float)

# 6. //=
a //= 2  # a = a // 2
print("a //= 2 →", a)  # Output: 3.0

# 7. **=
a **= 3  # a = a ** 3
print("a **= 3 →", a)  # Output: 27.0
```

- Comparison Operators (==, !=, >, <, >=, >=)


```python
x = 10
y = 5

print("x == y:", x == y)   # False
print("x != y:", x != y)   # True
print("x > y:", x > y)     # True
print("x < y:", x < y)     # False
print("x >= y:", x >= y)   # True
print("x <= y:", x <= y)   # False
print("x is y:", x is y)     # False
print("x is not y:", x is not y)  # True
```

- Logical Operators (and, or, not)

```python
# Logical Operators Examples

a = True
b = False

print("a and b:", a and b)   # False, because both are not True
print("a or b:", a or b)     # True, because at least one is True
print("not a:", not a)       # False, because a is True

# With numbers
x = 10
y = 5

print("(x > 5) and (y < 10):", (x > 5) and (y < 10))   # True
print("(x < 5) or (y < 10):", (x < 5) or (y < 10))     # True
print("not (x == y):", not (x == y))                   # True
```

- Membership Opertor (in, not in) ---> It checks, the value are present in data structure( list, tuple, string, set, dictionary) or not
    - in --> it return true when value are prsent in DS
    - not in --> it return true when value are not present in DS
    

```python
# Using 'in'
my_list = [1, 2, 3, 4]  # List
print(2 in my_list)      # True
print(5 in my_list)      # False

# Using 'not in'
my_string = "hello"     # String
print('h' not in my_string)  # False
print('z' not in my_string)  # True

# Tuple
nums = (1,2,3,4,5)
print(2 in nums)    # True
print(6 not in nums)    # False
```

- Identity Operator --> It check, both variables are points the same memory in objects or not.
    - is --> It return true when both variable point the same object
    - is not --> It return true when both variable points same object

```python
a = 5
b = 5
print(a == b)   # True (value are same)
print(a is b)   # True (same memory location because integer are cached)

x = [1, 2, 3]
y = [1, 2, 3]
z = x
print(x == y)
print(x is y)

print(id(x))    # memory location of x
print(id(z))    # memory location of z
print(id(y))    # memory location of y
```

- Bitwise Operator[& (AND), `(OR), ^ (XOR), ~ (NOT), << (Left Shift), >> (Right Shift)] ---> It works with binary numbers(0 and 1). In python, these operators are are apply only integer num, it direct operation on bits of number.

```python
a = 5       # Binary: 0101
b = 3       # Binary: 0011

print("a =", a, "→", bin(a))
print("b =", b, "→", bin(b))

# 1. Bitwise AND
print("a & b =", a & b)       # 0101 & 0011 = 0001 → 1

# 2. Bitwise OR
print("a | b =", a | b)       # 0101 | 0011 = 0111 → 7

# 3. Bitwise XOR
print("a ^ b =", a ^ b)       # 0101 ^ 0011 = 0110 → 6

# 4. Bitwise NOT (only on 'a')
print("~a =", ~a)             # ~0101 = -(a+1) = -6

# 5. Left Shift (a << 1)
print("a << 1 =", a << 1)     # 0101 << 1 = 1010 → 10

# 6. Right Shift (a >> 1)
print("a >> 1 =", a >> 1)     # 0101 >> 1 = 0010 → 2
```
