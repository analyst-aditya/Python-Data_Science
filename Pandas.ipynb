# __Pandas__
* In Python, Pandas is a open source data analysis and manipulation library which is use for structured data(tables, CSV files, databases)
* Pandas mainly use for data cleaning, analyzing and transforming the data.
* It is use in field of data science, machine learning and data analysis

👉 _In Simple words_: Pandas = Excel but more powerfull

```python
import pandas as pd
# Ek directory ko DataFrame me convert karna
data = {
    "name": ['a','b','c','d'],
    "age": [20,21,19,18],
    "address": ['BBK','LKO','GDA','KP']
}
df = pd.DataFrame(data)
df  # isko print ke andar bhi likh sakte hai lekin uske andar likhne par excel wala theme(mzaa) nahi ayega
```

<h2>Read Data</h2>
<p>
We use <code>pandas.read_csv()</code> function to read the csv file. In the bracket, we put the file path along with a quotation mark, so that pandas will read the file into a data frame from that address. The file path can be either an URL or your local file address.<br>
Because the data does not include headers, we can add an argument <code>headers = None</code>  inside the  <code>read_csv()</code> method, so that pandas will not automatically set the first row as a header.<br>
You can also assign the dataset to any variable you create.
</p>
