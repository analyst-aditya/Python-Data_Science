## For loop
- syntax----->

for variable in range(start, stop, incriment):
    # do something

```python
# For loop are used to iterate over a sequence (like a list, tuple, or string) or other iterable objects.
# It allows you to execute a block of code repeatedly for each item in the sequence.
# The syntax for a for loop in Python is:

iterable = [1, 2, 3, 4, 5]
# for item in iterable:
for item in iterable:
    # Code block to execute for each item
    print(item)
    
# for i in range(5):
#     print(i)
```

#### for loop with else

```python
l = [1,5,6,8]
for item in l:
    # print(item)
    print(item, end=" ")        # end ka use kisi bhi line ko horizontally print karne ke liye hota hai
else:       # this is use when you want to print something after completion of loop
    print("done")
```

```python
# structure ---> for variable in range (start, stop, increment)

for k in range(1, 21, 4):                   # in this line 1 is started point , 21 is range, and 4 is increment
    print(k)                # range() methode ka use kisi bhi range ke number ko print karne ke liye kiya jata hao starting point se lekar ending point tak
    
```

## while loop
- syntax --> 
    while(condition):
        # do something

```python
# exmample

i = 0
while(i<6): 
    print(i)
    i += 1
```

### break, continue statement

```python
# break statement

for i in range(11):
    if i == 5:
        break       # given condition tak code run hoga uske baad nahi 
    print(i)
    
print("loop ko chodh kar bahar aa gya")
```

```python
# continue statement

for i in range(11):
    if i == 4:
        print("skip the iteration")
        continue        # yaha par 4 ko skip kar dega
    print(i)
```

### pass keyword


```python
# pass statement ---> agar koi line apko run nahi karwana hai wha par pass keyword ka use kare isse wo line skip ho jati hai
for i in range(20):
    pass        # pass keyword --> ye line stop karo next line run karo

i = 0
while (i<10):
    print(i)
    i+=1
```
