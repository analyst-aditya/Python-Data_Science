# Matplotlib

###  Matplotlib Kya Hai?

- Ek open-source library hai.

- Mostly scientific aur data analysis me use hoti hai.

- Ye charts ko customize karne ke liye bohot options deti hai (colors, labels, legends, styles, etc.).

- Iske upar aur bhi libraries bani hain jaise Seaborn aur Pandas plotting.

## 🔹 Different Plots in Matplotlib

- Line Plot – plt.plot()

- Bar Chart – plt.bar()

- Histogram – plt.hist()

- Scatter Plot – plt.scatter()

- Pie Chart – plt.pie()

```python
import matplotlib.pyplot as plt

# Line Plot
x = [1, 2, 3, 4, 5]
y = [2, 4, 1, 8, 7]
plt.plot(x, y, color="red", marker="o")
plt.title("Line Plot Example")
plt.show()

# Bar Chart
plt.bar(x, y, color="blue")
plt.title("Bar Chart Example")
plt.show()

# Scatter Plot
plt.scatter(x, y, color="green")
plt.title("Scatter Plot Example")
plt.show()

# Pie Chart
values = [10, 20, 30, 40]
labels = ["A", "B", "C", "D"]
plt.pie(values, labels=labels, autopct="%1.1f%%")
plt.title("Pie Chart Example")
plt.show()
```

## 🔹Customization

- Line styles: linestyle='--', color='r', marker='o'

- Axis range: plt.xlim(), plt.ylim()

- Grid: plt.grid(True)

- Legends: plt.legend()

```python
plt.plot([1,2,3], [4,5,6], color="red", linestyle="--", marker="o", label="Line1")
plt.plot([1,2,3], [6,5,4], color="blue", label="Line2")
plt.legend()    # ye label show karta hai(eg: line1, line2)
plt.grid(True)  # isse background chart line(grids) show hoti hai
plt.show()
```

```python
x = [1,2,3,4]
sales = [10,20,25,40]
profit = [5,10,15,20]

plt.plot(x, sales, label = "Sales")
plt.plot(x, sales, label = "Profit")

plt.legend()
plt.show()
```

```python
# Grid add karna
plt.plot(x,sales)
plt.grid(True)

# Figure size change karna 
plt.figure(figsize=(4,2))
plt.plot(x,sales)
plt.show()
```

## 🔹Multiple Plots & Subplots

- Ek hi figure me multiple graphs: plt.subplot()

- Ek hi canvas me compare karne ke liye.

```python
plt.subplot(1,2,1)
plt.plot(x,sales)
plt.title("Sales")

plt.subplot(1,2,2)
plt.plot(x,profit)
plt.title("Profit")
plt.show()
```

```python
plt.subplot(1,2,1)   # 1 row, 2 column, first plot
plt.plot([1,2,3],[1,4,9], color="red", linestyle="--", marker="o", label="Line1")

plt.subplot(1,2,2)   # second plot
plt.plot([1,2,3],[1,2,3])
plt.show()
```

## 🔹Advanced Customization

- Figure size change karna: plt.figure(figsize=(8,5))

- Text annotate karna: plt.text() ya plt.annotate()

- Different color maps (cmap) use karna

- Styles apply karna: plt.style.use('ggplot'), plt.style.use('seaborn')

###  1. Figure Size Change Karna

By default figure chhoti hoti hai, isko bada/small karne ke liye:

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(4,2))   # width=8, height=5 (inches)
x = [1,2,3,4,5]
y = [10,20,25,30,40]

plt.plot(x, y, color="red")
plt.title("Figure Size Example")
plt.show()
```

### 2. Text aur Annotation Add Karna

Kabhi-kabhi graph par directly text ya arrow dikhana hota hai:

```python
x = [1,2,3,4,5]
y = [2,4,6,8,10]

plt.plot(x, y, marker="o")

# Simple text
plt.text(3, 6, "Mid Point", color="red")

# Annotation with arrow
plt.annotate("Highest Point", 
             xy=(5,10),        # point to mark
             xytext=(3.5,10),     # text position
             arrowprops=dict(facecolor='green', shrink=0.03))  
plt.show()
```

### 3. Color Maps (cmap)

Agar multiple points hai to unhe color scale se dikhane ke liye:

```python
import numpy as np

x = np.random.rand(50)
y = np.random.rand(50)
colors = np.random.rand(50)

plt.scatter(x, y, c = colors, cmap="viridis", s=100)
plt.colorbar()  # color scale bar show karega
plt.show()
```

### 4. Styles Use Karna

Matplotlib me predefined styles hote hain jisse graph aur beautiful lagta hai:

```python
plt.style.available   # sab styles ki list dikhata hai
```

```python
plt.style.use('tableau-colorblind10')   # style apply
x = [1,2,3,4,5]
y = [1,4,9,16,25]

plt.plot(x, y, marker="o")
plt.title("Styled Graph (ggplot)")
plt.show()
```

### 5. Multiple Customizations Together

```python
import matplotlib.pyplot as plt   # Fixed: Correct import (seeborn → matplotlib.pyplot)

plt.figure(figsize=(7,5))
plt.style.use("seaborn-v0_8")

x = [1,2,3,4,5]
y = [2,4,6,8,10]

plt.plot(x, y, marker="o", color="purple", linewidth=2, label="Line")
plt.title("Advanced Customization Example", fontsize=16, color="blue")
plt.xlabel("X-axis", fontsize=12)
plt.ylabel("Y-axis", fontsize=12)
plt.legend()
plt.grid(True, linestyle="--", alpha=0.8)
plt.show()
```

### 🔹 3D Plotting in Matplotlib

- 3D graphs banane ke liye hume mpl_toolkits.mplot3d use karna padta hai.

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np
```

### 1. 3D Axes Banana

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')      # iss se 3D axis create hota hai.
# aur 111 ka use karne se hum ek hi subplot create kar rahe hain. aur 121 se ham ek se zyada subplot create kar sakte hain.
# 👉 Isse ek 3D plotting space ban jata hai.
```

### 2. 3D Line Plot

```python
fig = plt.figure()      # isse ek naya figure create hota hai
ax = fig.add_subplot(111, projection='3d')

x = [1,2,3,4,5]
y = [2,4,6,8,10]
z = [10,20,25,30,40]

ax.plot(x, y, z, color="red", marker="o")
ax.set_title("3D Line Plot")
plt.show()
```

### 3. 3D Scatter Plot

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

x = np.random.rand(50)
y = np.random.rand(50)
z = np.random.rand(50)

ax.scatter(x, y, z, c=z, cmap="viridis", s=50)
ax.set_xlabel("X Axis")
ax.set_ylabel("Y Axis")
ax.set_zlabel("Z Axis")
plt.show()
# 👉 Yahaan cmap="viridis" color scale lagata hai.
```

### 4. 3D Surface Plot

- Surface plots mostly scientific data me use hote hain.

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

X = np.linspace(-5, 5, 100)
Y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(X, Y)
Z = np.sin(np.sqrt(X**2 + Y**2))

ax.plot_surface(X, Y, Z, cmap="plasma")
ax.set_title("3D Surface Plot")
plt.show()
# 👉 Isse ek smooth 3D surface milega jo sine wave dikhayega.
```

### 5. 3D Wireframe Plot

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

X = np.linspace(-5, 5, 30)
Y = np.linspace(-5, 5, 30)
X, Y = np.meshgrid(X, Y)
Z = np.sin(np.sqrt(X**2 + Y**2))

ax.plot_wireframe(X, Y, Z, color="black")
ax.set_title("3D Wireframe Plot")
plt.show()
```
