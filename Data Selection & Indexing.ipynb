## Selection in different ways
- row wise selection
    - iloc:  index based selection
    - loc : label based selection
- column wise selection
    - single column selection
        - dictionary like selection
        - object like selection
    - multiple like selection
        - dictionary like selection

```python
df.iloc[45]     # this is row based selection
```

```python
df.iloc[2:7]    # show table from 2 to 6
```

```python
df.iloc[[2,8,6,9]]      # isme hum specific rows ko select kar sakte hai
```

```python
# here is 'id' set as a index
df = df.set_index('id')
df
```

```python
df.loc[6414100192]  # isme hum index ke basis pe data ko select kar sakte hai , aur bracket me index value dena hota hai
```

```python
df.loc[[6414100192,263000018]]  # multiple index selection
```

```python
df.sort_index() # sort by index
```

```python
df.sort_index(ascending=False) # descending order me sort karne ke liye
```
